## Este notebook es para visualizar gr치ficamente un caso particular de la funci칩n que describe a una onda sinusoidal de fase cero:

## $y(t)=A sen(2 \pi ft) $

## Graficar con normalidad la funci칩n continua y graficar tal que se impriman muestras discretas de la onda sinusoidal usando el comando $subplot$.

## Define la amplitud a 5 unidades de distancia y la frecuencia a 50 unidades de frecuencia.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Vamos a trabajar con las unidades MKS.
t=np.linspace(0,10,100) #Definimos el tiempo como variable en un rango entre 0 y 10 segundos, por cuestiones
#del suavizado en la visualizaci칩n, vamos a poner que halla 100 unidades para graficar en este intervalo de tiempo.

A=5 #m
f=50 #Hz=1/s
onda_sinusoidal=A*np.sin(2*np.pi*f*t)
```

```python
from matplotlib import style
plt.figure(figsize = (16,8))
style.use('dark_background')

plt.subplot(2,1,1)
plt.plot(t,onda_sinusoidal)
plt.xlabel(r'$t$', fontsize = 20)
plt.ylabel(r'$y(t)$', fontsize = 20)

plt.subplot(2,1,2)
plt.stem(t,onda_sinusoidal) # muestras discretas
plt.xlabel(r'$t$', fontsize = 20)
plt.ylabel(r'$y(t)$', fontsize = 20)

plt.show()
```

## Bajo la misma estructura, ahora graficaremos cuatro funciones que describen cuatro ondas sinusoidales: $y_1(t)$, $y_2(t)$, $y_3(t)$, $y_4(t)$.

## Para lo anterior definimos:

## $y_1(t)=5sen(2 \pi 2t)$
## $y_2(t)=2sen(2 \pi 4t)$
## $y_3(t)=7sen(2 \pi 6t)$
## $y_4(t)=y_1(t)+y_2(t)+y_3(t)$

```python
t=np.arange(0,1,1/300)

y1=5*np.sin(2*np.pi*2*t)         

y2=2*np.sin(2*np.pi*4*t)         

y3=7*np.sin(2*np.pi*6*t)          

y4=y1+y2+y3                          
```

```python
plt.figure(figsize = (24,14)) # set the size of figure
plt.suptitle('Diferentes ondas sinusoidales', fontsize = 40) 

# Primera:
plt.subplot(2,2,1) 
plt.plot(t,y1,'r',linewidth = 3) 
plt.title('Primera onda sinusoidal', fontsize = 25)
plt.xlabel(r'$t$', fontsize = 20)
plt.ylabel(r'$y_1(t)$', fontsize = 20)

# Segunda:
plt.subplot(2,2,2)
plt.plot(t,y2,'y',linewidth = 3)
plt.title('Segunda onda sinusoidal', fontsize = 25)
plt.xlabel(r'$t$', fontsize = 20)
plt.ylabel(r'$y_2(t)$', fontsize = 20)

# Tercera:
plt.subplot(2,2,3)
plt.plot(t,y3,'g',linewidth = 3)
plt.title('Tercera onda sinusoidal', fontsize = 25)
plt.xlabel(r'$t$', fontsize = 20)
plt.ylabel(r'$y_3(t)$', fontsize = 20)

# Cuarta:
plt.subplot(2,2,4)
plt.plot(t,y4,'m',linewidth = 3)
plt.title('Cuarta onda sinusoidal', fontsize = 25)
plt.xlabel(r'$t$', fontsize = 20)
plt.ylabel(r'$y_4(t)$', fontsize = 20)

plt.show()
```
